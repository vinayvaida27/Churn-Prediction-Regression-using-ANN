```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing  import StandardScaler, OneHotEncoder, LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
print(data.head(5))
```

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
data
```

```python

from sklearn.preprocessing import  OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
Geo_transformed = onehot_encoder_geo.fit_transform(data[["Geography"]]).toarray()
Geo_transformed = pd.DataFrame(Geo_transformed, columns = onehot_encoder_geo.get_feature_names_out(['Geography']) )
Geo_transformed
```

```python
data
```

```python
from sklearn.preprocessing import  LabelEncoder
label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])
data = data.drop( ['Geography'], axis = 1)
data = pd.concat([data, Geo_transformed], axis =1)
data
```

```python
import pickle

## Save the Encoders and scaler

with open('label_reg.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)
    
with open('onehot_encoder_geo_reg.pkl' , 'wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
X = data.drop(['EstimatedSalary'], axis = 1)
y = data[['EstimatedSalary']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 41)
```

```python
scaler_reg = StandardScaler()

X_train = scaler_reg.fit_transform(X_train)
X_test = scaler_reg.transform(X_test)
y_train = scaler_reg.fit_transform(y_train)
y_test = scaler_reg.transform(y_test)

with open('scaler_reg1.pkl', 'wb') as file:
    pickle.dump(scaler_reg,file)
```

```python
X_train
```

## Ann Regression Problem Statement

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense       
```

```python
model = Sequential (
    [
        
        Dense(64, activation='relu', input_shape = (X_train.shape[1],)),
        Dense(32, activation = 'relu'),
        Dense(1)
    ]
)

model.compile( optimizer = 'adam', loss = 'mean_absolute_error', metrics =['mae'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime

# Set up Tensor Board

log_dir = "regressionlogs/fit/" + datetime.datetime.now().strftime("%Y%m%d - %H%M%S")
tensorboard_callback = TensorBoard(log_dir = log_dir, histogram_freq=1)
# set up Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience = 10, restore_best_weights=True)
```

```python


history = model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test),
    epochs=100,
    callbacks= [early_stopping_callback, tensorboard_callback]
)
```

```python
test_loss, test_mae = model.evaluate(X_test, y_test)

test_loss
```
